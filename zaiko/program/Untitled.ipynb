```python
10分のズレは許容範囲です。
「検収日時」を昇順に並べたときに「順立装置入庫日時」が昇順になっていないかんばんの合計数を数えて

10分の許容範囲を考慮して計算した結果、合計1,881枚/26301のかんばん
```

```python
#ライブラリのimport
import pandas as pd
import numpy as np
%matplotlib inline
#from matplotlib import pyplot as pyp
import matplotlib.pyplot as plt
```

```python
pre_zaikodata = pd.read_csv('..//data//設計値変更後_LTデータ9月（土日削除）.csv',encoding='shift_jis')
pre_zaikodata.head(300)
```

```python
count = 0
sedata = pd.read_excel('..//data//仕入れ先便.xlsx',encoding='cp932')
long2 = sedata.iloc[:,1]
long = pre_zaikodata.iloc[:,1]

for i in range(len(long)):
    i
    for j in range(len(long2)):
        if (pre_zaikodata.loc[i,"仕入先名"]==sedata.loc[j,"仕入先名"]) and (pre_zaikodata.loc[i,"仕入先工場名"]==sedata.loc[j,"発送場所名"]):
            index = sedata.columns.get_loc("納入先") + int(pre_zaikodata.loc[i,"納入便"])
            pre_zaikodata.loc[i,"納入時間"]=sedata.iloc[j,index]
```

```python
with open('..//data//設計値変更後_LTデータ9月（土日削除）納入時間付き.csv', mode='w',newline='', encoding='shift_jis',errors='ignore') as f:
    pre_zaikodata.to_csv(f)
```
